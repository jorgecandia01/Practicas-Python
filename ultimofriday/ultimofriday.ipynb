```python
goodf=[]
with open('DatasaurusDozen.tsv','r') as file:
    f=file.readlines()
    for i in f:
        goodf.append(i[0:-1])
```

```python
dataset=[]
x=[]
y=[]
for i in goodf:
    a=i.split('\t')
    dataset.append(a[0])
    x.append(a[1])
    y.append(a[2])
    
count=0
for i in x:
    if count!=0:
        x[count]=float(i)
    count+=1

count=0
for i in y:
    if count!=0:
        x[count]=float(i)
    count+=1
```

```python
roots=[]
terms=[]
count=0
for i in dataset:
    if count!=0:
        if dataset[count]!=dataset[count-1]:
            roots.append(count)
            terms.append(i)
        
    count+=1
roots.append(len(x))
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(13,1,figsize=(5,50))
for i in range(13):
    axs.flat[i].scatter(x[roots[i]:roots[i+1]],y[roots[i]:roots[i+1]])
```
