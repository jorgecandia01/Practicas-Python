## Lab 10.2. Third grade function

```python
import numpy as np
import matplotlib.pyplot as plt  

x = np.linspace(-5, 5, 1000)

y = 0.25 *(x+4)* (x+2) * (x-3)
```

```python
import matplotlib.lines as mlines

#Print function
plt.plot(x,y, linewidth=3)
plt.axis ([-8,8,-20,35])


#Required to put 6 to obtain a list from -5 to 5
listx = range(-5,6)
myticks=np.array(listx)
#alternative: np.arange(-5,6)
#set x axis values
plt.xticks(myticks)


#Print green axis
x2= np.linspace(-8, 8, 1000)
y2=x2*0
plt.plot (x2,y2,'g--')

y3= np.linspace(-20, 35, 1000)
x3= y3*0
plt.plot (x3,y3,'g--')

#print roots
list = [-4,-2,3]
rootsx=np.array(list)
rootsy = rootsx *0
plt.plot (rootsx, rootsy, 'r*', markersize=10)

#print annotations
plt.annotate('root 1', xy=(-4,0), xytext=(-6,20), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('root 2', xy=(-2,0), xytext=(.2,15), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('root 3', xy=(3,0), xytext=(6,-10), arrowprops=dict(facecolor='black', shrink=0.05))


# Hint: to print the * in the legend we change the color of the line to white
funct_line = mlines.Line2D([], [], color='blue', label='Function: y = 0.25 *(x+4)* (x+2) * (x-3)')
axis_line = mlines.Line2D([], [], color='green', linestyle='--',  label='Axis')
root_dots = mlines.Line2D([], [], color='w',markerfacecolor='red', marker='*',markersize=15, label='Roots')

plt.legend(handles=[funct_line, axis_line, root_dots],bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

## Lab 10.3. Subplotting Sin function

```python

from math import sin
from math import pi

items=input("Introduce number of charts: ")
items=int(items)

while items % 2 !=0:
    print("\nIntroduced value is not an even number")
    print("Please introduce a new value")
    items=input("Introduce number of charts: ")
    items=int(items)

# It is mandatory to have two columns in the subplot. To obtain the number of rows we divide 
# items by 2

cols=2
rows= int(items/2)

fig, axs = plt.subplots(rows, cols)
plt.subplots_adjust(top=1.2, bottom=0.0001, left=0.10, right=0.95, hspace=0.5,wspace=0.35)


ang= np.linspace(-2* pi, 2* pi, 1000)

factor=1

for i in range(0, rows):
    for j in range(0, cols):
        
        #si no se usan las funciones de numoy uede dar problemas
        y =np.sin(ang * factor)
        axs[i, j].plot(ang, y)
        title= ('Image' + str(factor) + ': sin(' + str(factor) +'x)' + ' Subplot('+ str(i)+ ',' + str(j)+ ')' )
        axs[i, j].set_title(title)
        factor+=1
    
```

## Lab 10.4. Plot 3D

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# Make data.
x= np.linspace(0, 2* pi, 500)
y =x

xx, yy = np.meshgrid(x, y, sparse=True)

z = np.sin(xx)+ np.cos(yy)

# Plot the surface.
surf = ax.plot_surface(xx, yy, z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis.
#ax.set_zlim(-1.01, 1.01)
#ax.zaxis.set_major_locator(LinearLocator(10))
#ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Lab 10.5. Anscombe's quartet 

### Anscombe's quartet comprises four data sets that have nearly identical simple descriptive statistics, yet have very different distributions and appear very different when graphed. Each dataset consists of eleven (x,y) points. They were constructed in 1973 by the statistician Francis Anscombe to demonstrate both the importance of graphing data before analyzing it 

```python
import matplotlib.pyplot as plt
import numpy as np

x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,10))


axs[0, 0].plot(x, y1, 'ks')
axs[0, 0].axis([2, 20, 2, 14])
axs[0, 0].text(17, 12, 'I', fontsize=20)

axs[0, 1].plot(x, y2, 'ks')
axs[0, 1].axis([2, 20, 2, 14])
axs[0, 1].text(17, 12, 'II', fontsize=20)

axs[1, 0].plot(x, y3, 'ks')
axs[1, 0].axis([2, 20, 2, 14])
axs[1, 0].text(17, 12, 'III', fontsize=20)

axs[1, 1].plot(x4, y4, 'ks')
axs[1, 1].axis([2, 20, 2, 14])
axs[1, 1].text(17, 12, 'IV', fontsize=20)
,

pairs = (x, y1), (x, y2), (x, y3), (x4, y4)
for x, y in pairs:
    print('mean=%1.2f, std=%1.2f, r=%1.2f' % (np.mean(y), np.std(y),np.corrcoef(x, y)[0][1]))


plt.show()
```

## Lab 10.6. Bicimad

```python
import pandas as pd

df=pd.read_csv('bicimad.csv',sep=";")
df.head()
```

### Histograme of user's ages

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,6))
plt.hist(df['ageRange'], range=[-0.5,6.5], bins=7, align="mid", rwidth=0.5)
plt.ylabel('Number of travels');
plt.xlabel('Groups of age');

line1 = mlines.Line2D([], [], color='blue', label='0: No se ha podido determinar el rango de edad del usuario')
line2 = mlines.Line2D([], [], color='blue', label='1: El usuario tiene entre 0 y 16 aos')
line3 = mlines.Line2D([], [], color='blue', label='2: El usuario tiene entre 17 y 18 aos')
line4 = mlines.Line2D([], [], color='blue', label='3: El usuario tiene entre 19 y 26 aos')
line5 = mlines.Line2D([], [], color='blue', label='4: El usuario tiene entre 27 y 40 aos')
line6 = mlines.Line2D([], [], color='blue', label='5: El usuario tiene entre 41 y 65 aos')
line7 = mlines.Line2D([], [], color='blue', label='6: El usuario tiene 66 aos o ms')

plt.legend(handles=[line1, line2, line3, line4, line5, line6, line7],bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

### Histogram of travel times

If we plot in log scale, we can see that there are 4 orders of magnitude of travels with duration > 5000. Those must be outliers

```python
plt.hist(df['travel_time'])
plt.ylabel('Number of travels');
plt.xlabel('Time_travel range in seconds');
plt.yscale("log")
```

Let's plot only trips on a reasonable range, between 0 and 50 minutes (3000 seconds)

```python
plt.figure(figsize=(8,6))
plt.hist(df['travel_time'],bins=50,range=[0,3000])
plt.title("Histogram of trip times")
plt.ylabel('Number of travels');
plt.xlabel('Time_travel range in seconds');
plt.show()
```

### Number of trips per days

```python
fig=plt.figure(figsize=(15,5))
serie=df.groupby(["Fecha"])["ageRange"].count()
plt.bar(serie.index,serie)
fig.autofmt_xdate()
plt.title("Frecuency of trips per day")
plt.ylabel('Number of travels');
plt.xlabel('Day');
plt.show()
```
