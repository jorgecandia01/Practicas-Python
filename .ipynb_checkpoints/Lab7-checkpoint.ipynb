# Lab 7.4 Get words from a long file (2 pts.)

```python
with open("ElQuijote.txt","r",encoding="utf8") as q:
    textoentero=q.readlines()
```

```python
lineas=[]
palabras=[]
for i in textoentero:
    linea=i.split("\n")
    for palabra in linea:
        lineas.append(palabra)

for i in lineas:
    palabrejas=i.split(" ")
    for palabreja in palabrejas:
        palabras.append(palabreja)
```

```python
caracteresraros= """,;:.-_{}()"?¿!'\@#$%&/[]"""
caracs=[]
for i in caracteresraros:
    caracs.append(i)
    
print(caracs)
```

```python
for carac in caracs:
    for palabra in palabras:
        if carac in palabra:
            palabrejas=palabra.split(carac)
            palabras.remove(palabra)
            for palabreja in palabrejas:
                palabras.append(palabreja)
                
for carac in caracs:
    for palabra in palabras:
        if carac in palabra:
            palabrejas=palabra.split(carac)
            palabras.remove(palabra)
            for palabreja in palabrejas:
                palabras.append(palabreja)
```

```python
while "" in palabras:
    palabras.remove("")
```

```python
lista_palabras_en_minuscula=[]
for palabra in palabras:
    pminuscula=palabra.lower()
    lista_palabras_en_minuscula.append(pminuscula)
    

    
#lista_palabras_en_minuscula      variable con todas las palabras en minuscula
#palabras                         variable con todas las palabras como estan en el texto
```

# Lab 7.6 Frequency of words (1 pt.)

```python
frecuencia={}
for i in lista_palabras_en_minuscula:
    if i in frecuencia.keys():
        frecuencia[i]=frecuencia[i]+1
    else:
        frecuencia[i]=1
frecuencia
```

```python
import pickle
with open("palabrasquijote.dat","wb") as quij:
    pickle.dump(frecuencia,quij)
```

# Lab 7.7 Read binary files (2 pts.)

```python
import pickle
quij=open("palabrasquijote.dat","rb")
frec=pickle.load(quij)
quij.close()
```

```python
items=[]
for item in frec.items():
    listilla=[]
    for i in item:
        listilla.append(i)
        listilla.reverse()
    items.append(listilla)
    

items.sort()
items.reverse()
```
