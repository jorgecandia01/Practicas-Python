# Lab 10.2. Third grade function (2 pt.)


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(-5,5)
x=np.linspace(-5,5,1000)
y = 0.25 *(x+4)* (x+2) * (x-3)

ceros=[]
for i in range(200):
    ceros.append(0) 
    
casiroots=[]
i=0
for valor in y:
    if valor>=0 and y[i-1]<=0 or valor<=0 and y[i-1]>=0 and valor!=y[len(y)-1] and valor!=y[0]:
        casiroots.append(x[i])
    i+=1
    
    
plt.plot(x,y,label='Function: y = 0.25 *(x+4)* (x+2) * (x-3)')
plt.plot(range(-100,100),ceros,'g--',label='Axis')
plt.plot(ceros,range(-100,100),'g--')
plt.plot(casiroots,[0,0,0],'*r',label='Roots')
plt.annotate('Root 1', xy=(casiroots[0],1), xytext=(-6,15),arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Root 2', xy=(casiroots[1],1), xytext=(-2,25),arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Root 3', xy=(casiroots[2],1), xytext=(6,-9),arrowprops=dict(facecolor='black', shrink=0.1))


plt.legend(bbox_to_anchor=(1.02,1), loc='upper left')
plt.axis([-9,9,-25,35])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
with open('bicimad.csv','r') as file:
    f=file.readlines()
    del f[0]
```

```python
fecha=[]
ageRange=[]
user_type=[]
idunplug_station=[]
idplug_station=[]
idunplug_base=[]
idplug_base=[]
travel_time=[]
fichero=[]
for apartado in f:
    apartados=apartado.split(';')
    fecha.append(apartados[0])
    ageRange.append(apartados[1])
    user_type.append(apartados[2])
    idunplug_station.append(apartados[3])
    idplug_station.append(apartados[4])
    idunplug_base.append(apartados[5])
    idplug_base.append(apartados[6])
    travel_time.append(apartados[7])
    fichero.append(apartados[8])
```

```python
ageRange.sort()
counter=[]
for i in range(int(ageRange[-1])+1):
    counter.append(ageRange.count(str(i)))
counter
```

```python
count=0
edades=['No se ha podido determinar la edad del usuario',
        'El usuario tiene entre 0 y 16 aos',
        'El usuario tiene entre 17 y 18 aos',
        'El usuario tiene entre 19 y 26 aos',
        'El usuario tiene entre 27 y 40 aos',
        'El usuario tiene entre 41 y 65 aos',
        'El usuario tiene 66 aos o ms']
for i in range(len(counter)):
    plt.bar(count,counter[i],label=str(count)+': '+edades[count])
    plt.legend()
    plt.legend(bbox_to_anchor=(1.02,1.022))
    count+=1
```

```python
semana=[]
semanita=['31','01','02','03','04','05','06']
dias=[]
for i in fecha:
    if i[:2] in semanita:
        semana.append(i)

fakesemana=semana[:]     
fakesemana.sort()
for i in range(7):
    a=fakesemana[0]
    dias.append(fakesemana.count(fakesemana[0]))
    
    for i in fakesemana:
        if fakesemana[0]==a:
            del fakesemana[0]
            
```
