```python
days=[]
fakemaxs=[]
Tmin=[]
Tmax=[]
with open("lab8excel.csv",'r') as f:
    #text=f.read().split("\n")
    #f.seek(0)
    header=f.readline()
    print(header)
    #while '' in text:
     #   text.remove('')
    for line in range(7):
        a=f.readline().split(';')
        days.append(a[0])
        Tmin.append(a[1])
        fakemaxs.append(a[2])
    for i in fakemaxs:
        ii=i.split("\n")
        Tmax.append(ii[0])
print(days)
print(Tmin)
print(Tmax)

iTmax=[]
iTmin=[]
for i in Tmax:
    iTmax.append(int(i))
for i in Tmin:
    iTmin.append(int(i))
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.plot(days,iTmax,'y')
plt.plot(days,iTmin,'c')
plt.xticks(range(7),days,rotation='vertical')
```
